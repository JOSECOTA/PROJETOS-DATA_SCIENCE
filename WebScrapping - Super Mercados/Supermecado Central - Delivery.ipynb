```python
import bs4
import urllib.request as urllib_request
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
def lista_produtos(produto):
    url = 'https://delivery.supercentral.com.br/loja/busca/?q='+produto+'&page=1'
    response = urlopen(url)
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')
    num_itens = int((soup.h1.get_text()).split()[0])
    if num_itens % 20 == 0:
        num_paginas = num_itens/20
    else:
        num_paginas = int(num_itens/20)+1

    lista_nome = []
    lista_preco = []

    for a in range(num_paginas):
        url = 'https://delivery.supercentral.com.br/loja/busca/?q='+produto+'&page='+str(a+1)
        response = urlopen(url)
        html = response.read()
        soup = BeautifulSoup(html, 'html.parser')
        for i in range(len(soup.findAll('span',{'class':'preco-final'}))):
            lista_nome.append(soup.findAll('a',{'class': 'link-nome-produto'})[i].get_text().replace('\n',''))
            lista_preco.append(soup.findAll('span',{'class':'preco-final'})[i].get_text().replace(',','.'))

    central = pd.DataFrame({'Supermercado':'Central','Produto':lista_nome , 'Preço':lista_preco})
    central['Preço'] = central['Preço'].astype(float)
    
    
    return central.sort_values('Preço').reset_index(drop=True)




    
```

```python
lista_produtos('skol')
```
